```python
"""Considere um arquivo CSV (comma separated values) com os seguintes com
dados de todas as copas do mundo até o ano de 2014
• As informações do arquivo são
• Year - Ano da copa
• Country - Pais sede
• Winner - Campeão
• Runners-Up - Vice-Campeão
• Third - Terceiro lugar
• Fourth - Quarto Lugar
• GoalsScored - Total de Gols
• QualifiedTeams - Quantidade de Times
• MatchesPlayed - Quantidade de Jogos
• Attendance - Publico Total

Pede-se
• Crie uma função que efetue a leitura do arquivo e crie um arquivo de saida com as
seguintes informações (uma informação por linha):
• Soma de público das copas com anos final 0 (1930, 1950, etc)
• Quantidade total de gols entre as copas de 1954 e 1990, inclusive
• Média de público
• Média de gols por partida
• Quantidade de vezes em que o país sede foi campeão
• Quantidade de vezes em que o time do Brasil ficou entre uma das 4 primeiras posições
• Ano das edições em que o time da França finalizou em terceiro lugar
• Quantidade de vitórias por país, classificada em ordem crescente do número de títulos

"""
```

```python
def worldCup(fileName):
    file = open (fileName)
    linecount = 0
    somaPublico = 0
    GoalsTotal = 0
    mediaPublico = 0
    qntCopas = 0
    sedeCampeao = 0
    MediaGols = 0
    MediaGolsGoalsScored = 0
    MediaGolsMatchesPlayed = 0
    somaBrasilQuali = 0
    somaFrancaQuali = 0
    campeoes = []
    
    for line in file:
        if(linecount>0):
            lineSplit = line.split(',')
            year = lineSplit[0]
            Country = lineSplit[1]
            Winner = lineSplit[2]
            RunnersUp = lineSplit[3]
            Third =  lineSplit[4]
            Fourth = lineSplit[5]
            GoalsScored = lineSplit[6]
            QualifiedTeams = lineSplit[7]
            MatchesPlayed = lineSplit[8]
            Attendance = lineSplit[9]
    
            
            if(year.endswith('0')):
                somaPublico += int(Attendance.replace('\n', '').replace('.', ''))
            
           
            if(int(year) >= 1954 and int(year)<=1990):
                GoalsTotal += int(GoalsScored)
            
            #Média de público
            mediaPublico+= int(Attendance.replace('\n', '').replace('.', '')) 
            qntCopas+=1
            
            #Quantidade de vezes em que o país sede foi campeão
            if(Country == Winner):
                sedeCampeao += 1
            
            #Média de gols por partida
            MediaGolsGoalsScored += int(GoalsScored)
            MediaGolsMatchesPlayed =+ int(MatchesPlayed)
            
            #Quantidade de vezes em que o time do Brasil ficou entre uma das 4 primeiras posições
            if( Winner=='Brazil' or RunnersUp=='Brazil' or Third=='Brazil' or Fourth=='Brazil'):
                somaBrasilQuali += 1
            
            # Ano das edições em que o time da França finalizou em terceiro lugar
            if(Third=='France'):
                somaFrancaQuali += 1
            
            #Quantidade de vitórias por país, classificada em ordem crescente do número de títulos  
            campeoes.append(Winner)
            
        else:
            linecount +=1; 
    
    
    
    
    
    
    
    print ('Publico', somaPublico)
    print ('Gols', GoalsTotal)
    print ('mediaPublico', mediaPublico / qntCopas)
    print ('país sede foi campeão', sedeCampeao)
    print ('Media de Gols - MediaGolsMatchesPlayed', MediaGolsMatchesPlayed)
    print ('Media de Gols - MediaGolsGoalsScored',MediaGolsGoalsScored)
    print ('Media de Gols', round(MediaGolsGoalsScored / MediaGolsMatchesPlayed , 2))
    print ('Brasil ficou entre uma das 4 primeiras posições', somaBrasilQuali)
    print ('França finalizou em terceiro lugar', somaFrancaQuali)
    
    mysetCampeoes = set(campeoes)
    listaOrdenada = []
    for time in mysetCampeoes:
        listaOrdenada.append([time, campeoes.count(time)])

    print ('Quantidade de vitórias por país', sorted(listaOrdenada, key=lambda listaOrdenada: listaOrdenada[1]) )
    
```

```python
worldCup('WorldCups.csv')
```
