```python
'''• Escreva um programa que calcule os valores das tabuadas do 1 ao 10 de forma simultânea (1
thread por tabuada)
• Imprima na tela a lista com os resultados de cada tabuada
```

```python
import threading, time, random


class TimesTable(threading.Thread):
    def __init__(self, number, ret_list):
        threading.Thread.__init__(self)
        self.number = number
        self.ret_list = ret_list

    def run(self):
        print('Starting Thread: ', self.getName())
        for i in range(1, 11):
            print('Thread: ', self.getName(), ' calculating...')
            self.ret_list.append(i * self.number)


threads = []
[threads.append(TimesTable(i, [])) for i in range(1, 11)]
[t.start() for t in threads]
[t.join() for t in threads]
[print(t.ret_list) for t in threads]
```
