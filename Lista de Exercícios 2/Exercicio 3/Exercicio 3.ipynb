```python
'''Considere um arquivo de entrada no formato CSV (comma separated
values) com informações relativas gastos de parlamentares do estado de
São Paulo, entre os anos de 2009 e 2018 (cota_parlamentar_sp.csv)
• As informações estão separadas pelo caracter separador ; conforme o
exemplo abaixo:
01/01/2018 00:00;3011;2015;2018;1;PPS;ALEX MANENTE;MANUTENÇÃO DE ESCRITÓRIO
DE APOIO À ATIVIDADE PARLAMENTAR;;CENTRO EMPRESARIAL MEDITERRANEO;91606
01/01/2018 00:00;3011;2015;2018;1;PPS;ALEX MANENTE;MANUTENÇÃO DE ESCRITÓRIO
DE APOIO À ATIVIDADE PARLAMENTAR;;CENTRO EMPRESARIAL MEDITERRANEO;86864

Crie um programa que efetue a leitura do
arquivo (cota_parlamentar_sp.csv) e para
cada legenda (sgpartido) grave um
arquivo com o nome da legenda no
formato JSON (ex: DEM.json, PSDB.json)
• Em cada arquivo deverá conter as
seguintes informações conforme exemplo
ao lado:
• legenda (partido)
• nome (nome do parlamentar)
• total (total de despesas do parlamentar)
```

```python
import json
```

```python
def readFile(fileName):
    f = open (fileName, encoding = "ISO-8859-1").readlines() #lendo o arquivo
    header = f.pop(0) #guardando o cabeçalho e removendo ele da lista
    lineArray = [] #Array pra guardar linha
    tableArray = [] #array pra guardar tabela toda
    for line in f:
        splitLine = line.split(';') #separando as celulas por split ~
        for item in splitLine:
            item = item.replace("\"", "") #removendo o aspas
            item = item.replace("\n", "") #removendo o \n            
            lineArray.append(item) #adicionando a celula no array da linha
        tableArray.append(lineArray) #adicionando a linha na tabela
        lineArray=[] #limpando o array da linha para ser usado na proxima interação
    return tableArray
```

```python
def createJsonFile(jsonContent, JSONFileName):
    estatistics = []
    
    #criando o conteudo para o arquivo JSON
    for item in jsonContent:
        estatistics.append({
            'total': item[1],
            'nome': item[0],
            'legenda':JSONFileName
        })
   
    #print(estatistics)
    
    #Salvando estatisticas no arquivo JSON
    with open(JSONFileName + '.json', 'w') as outfile:
       json.dump(estatistics, outfile,indent=4,)
```

```python
#Esta função cria um array bidimensional
#Na Primeira dimensão guarda o nome do partido
#Na segunda uma lista com os todos valores gastos por cada politico 
def createJsonContent(tableArray):
    jsonBodyArray = []
    for line in tableArray:
        partido = line[5] #Guardando o nome do arquivo
        x=[] 
        try:
            x = [x for x in jsonBodyArray if partido in x][0] #Validando se o partido existe na primeira dimensao
        except:
             a= 0
            
    
        # Se a fase operação nao foi ele localizado inclui a fase operação no array
        if len(x) == 0:
            # Se o partido ainda não existir no array, 
            # cria-se ele e cria um array vazio na segunda dimensão
            jsonBodyArray.insert(len(jsonBodyArray)+1 , [partido, []])  
 
        x = [x for x in jsonBodyArray if partido in x][0] # localizando o partido no bidimensional    
        indexAFO = jsonBodyArray.index(x) #pegando o indice desse partido
        
        #Insere o nome do candidato e o valor gasto por ele na segunda dimensão
        jsonBodyArray[indexAFO][1].insert(len(jsonBodyArray[indexAFO][1])+1, [line[6], line[10]]) 
    
    
    #Depois de ler todo o arquivo e criar o array bidimensional, vou interar nele para gerar os arquivos .JSON
    #A ideia aqui é criar um novo array apenas com o nome unico de cada candidato e  ir somando os gastos que ele teve
    
    listaPoliticoGastos = []        
    for item in jsonBodyArray:
        totalPoliticoPartido = []
        partido = item[0]

        for partidoTotal in item[1]:
            
            x=[]
            try:
                x = [x for x in listaPoliticoGastos if partidoTotal[0] in x][0] # localizando o item no array bidimensional
            except:
                a= 0

            if len(x) == 0:
                listaPoliticoGastos.insert(len(listaPoliticoGastos)+1, [partidoTotal[0], 0])
            
            
            #Busca na lista o nome do candidato
            #pega o seu indice
            #soma-se o valor anterior ao valor atual do gasto
            #atualiza p total gasto pelo candidato
            x = [x for x in listaPoliticoGastos if partidoTotal[0] in x][0]
            indexCandidato = listaPoliticoGastos.index(x)
            totalCandidato = int(listaPoliticoGastos[indexCandidato][1]) + int(partidoTotal[1])
            listaPoliticoGastos[indexCandidato][1] = totalCandidato 
        
        #no final da leitura de cada partido, se cria o arquivo JSON
        createJsonFile(listaPoliticoGastos, partido)

    return jsonBodyArray
```
